```python
# Importing all necessary libraries
import cv2
import os
  
def get_frames(video_name, ext):
    # Read the video from specified path
    cam = cv2.VideoCapture(f"{video_name}.{ext}")

    try:

        # creating a folder named data
        if not os.path.exists('video_images'):
            os.makedirs('video_images')

    # if not created then raise error
    except OSError:
        print ('Error: Creating directory of data')

    # frame
    currentframe = 0

    while(True):

        # reading from frame
        ret,frame = cam.read()

        if ret:
            # if video is still left continue creating images
            name = f'./video_images/{video_name}_frame_{currentframe:03}.jpg'
            print ('Creating...' + name)

            # writing the extracted images
            cv2.imwrite(name, frame)

            # increasing counter so that it will
            # show how many frames are created
            currentframe += 1
        else:
            break

    # Release all space and windows once done
    cam.release()
    cv2.destroyAllWindows()
```

```python
get_frames('MVI_5959','MP4')
```
