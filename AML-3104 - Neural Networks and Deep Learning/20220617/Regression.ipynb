```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split # splitting the data
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.preprocessing import StandardScaler # data normalization
from sklearn.metrics import confusion_matrix
from sklearn import metrics  
import itertools
import matplotlib.pyplot as plt
```

```python
def train_regressor(result_list, reg_type, features, regressor, x_var, y_var, binary_func = None):
    x_train, x_test, y_train, y_test = train_test_split(x_var, y_var, test_size = 0.3, random_state = 4)
    # Uses StandardScaler to normalize the data
    x_train = StandardScaler().fit_transform(x_train)
    x_test = StandardScaler().fit_transform(x_test)
    # Sets up the regression using the training data
    regressor.fit(x_train, y_train)
    # Predicts the values
    y_pred = regressor.predict(x_test)
    # If a lambda function is provided then call it to obtain the y values
    if binary_func:
        y_pred = binary_func(y_pred)
    # Calculates the confusion Matrix
    conf_m = confusion_matrix(y_test, y_pred, labels=[1,0])
    # Appends the results to the list
    result_list.append({
        'regressor': reg_type,
        'features': f'{features}',
        'feature_count': len(features),
        'confusion_matrix': f'{conf_m}',
        'r2_score': metrics.r2_score(y_test, y_pred),
        'mean_absolute_error':metrics.mean_absolute_error(y_test, y_pred),  
        'mean_squared_error': metrics.mean_squared_error(y_test, y_pred),  
        'regressor_instance': regressor,
        'data': {
            'x_test': x_test,
            'y_test': y_test,
            'y_pred': y_pred
        }
    })
    
```

```python
def test_feature_sets(result_list, df, field_set, dependant_variable):
    #Split the data
    x_var = np.asarray(df[field_set])
    y_var = np.asarray(df[dependant_variable])
    
    try:
        # Logistic Regression
        log_regressor = LogisticRegression(C = 0.1, solver = 'liblinear')
        train_regressor(result_list, 'Logistic', field_set, log_regressor, x_var, y_var)
    except Exception as e:
        print(f'!! An exception has ocurred: {e}')
    
    try:
        # Linear Regression
        linear_regressor = LinearRegression()
        train_regressor(result_list, 'Linear', field_set, linear_regressor, x_var, y_var, lambda y: [round(v) for v in y])
    except Exception as e:
        print(f'!! An exception has ocurred: {e}')
    
    
```

# Columns of the dataset
- tp: Number of times pregnant
- pgc: Plasma glucose concentration a 2 hours in an oral glucose tolerance test
- dbp: Diastolic blood pressure (mm Hg)
- thsi: Triceps skin fold thickness (mm)
- thsi: 2-Hour serum insulin (mu U/ml)
- bmi: Body mass index (weight in kg/(height in m)^2)
- dpf: Diabetes pedigree function
- age: Age (years)

```python
df = pd.read_csv('pima-indians-diabetes.csv', names = ['tp', 'pgc', 'dbp', 'tsft', 'thsi', 'bmi', 'dpf', 'age', 'class'])
df.head(5)
```

```python
all_features = ['tp', 'pgc', 'dbp', 'tsft', 'thsi', 'bmi', 'dpf', 'age']
result_list =  []
# Tries all the combinations of features from Lenght = 1 up to n
for L in range(0, len(all_features)):
    # Gets all the combinations for a given lenght
    for subset in itertools.combinations(all_features, L+1):
        test_feature_sets(result_list, df, list(subset), 'class')
analysis_df = pd.DataFrame(result_list)
```

```python
best_feature_sets_df = analysis_df.sort_values(['r2_score', 'feature_count'], ascending=[False, True])
best_feature_sets_df.head(10)
```

```python
best_regressor = best_feature_sets_df['regressor_instance'].iloc[0]
best_data = best_feature_sets_df['data'].iloc[0]
```

```python
plt.plot(best_data['x_test'], best_data['y_test'], 'o', color='red')
plt.plot(best_data['x_test'], best_data['y_pred'], 'x', color='blue')
```
