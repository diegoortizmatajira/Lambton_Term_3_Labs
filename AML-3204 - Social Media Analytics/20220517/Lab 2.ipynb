```python
import pandas as pd
import re
import csv
```

```python
data_lines = []
with open("YouTubeTrailerComments.txt") as f:
    lines = f.readlines()
    id = 0
    for line in lines:
        if line.startswith('MovieID::::'):
            id = int(line.replace('MovieID::::', ''))
        elif re.match('^\d+(\s\d+){4}$', line) != None:
            data_lines.append(f"{id},{line.replace(' ',',')}")
```

```python
reader = csv.reader(data_lines)
parsed_csv = list(reader)
```

```python
df = pd.DataFrame(parsed_csv, columns = ['id','vc','lc','dc','fc','cc'])
```
